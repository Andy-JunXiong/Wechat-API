```python
import itchat

# Ëé∑ÂèñÊï∞ÊçÆ
def get_data():
    # Êâ´Êèè‰∫åÁª¥Á†ÅÁôªÈôÜÂæÆ‰ø°ÔºåÂÆûÈôÖ‰∏äÂ∞±ÊòØÈÄöËøáÁΩëÈ°µÁâàÂæÆ‰ø°ÁôªÈôÜ
    itchat.auto_login()
    # Ëé∑ÂèñÊâÄÊúâÂ•ΩÂèã‰ø°ÊÅØ
    friends = itchat.get_friends(update=True)  # ËøîÂõû‰∏Ä‰∏™ÂåÖÂê´Áî®Êà∑‰ø°ÊÅØÂ≠óÂÖ∏ÁöÑÂàóË°®
    return friends

if __name__ == '__main__':
    print(get_data())
```

```python
def parse_data(data):
    friends = []
    for item in data[1:]:  # Á¨¨‰∏Ä‰∏™ÂÖÉÁ¥†ÊòØËá™Â∑±ÔºåÊéíÈô§Êéâ
        friend = {
            'NickName': item['NickName'],  # ÊòµÁß∞
            'RemarkName': item['RemarkName'],  # Â§áÊ≥®Âêç
            'Sex': item['Sex'],  # ÊÄßÂà´Ôºö1Áî∑Ôºå2Â•≥Ôºå0Êú™ËÆæÁΩÆ
            'Province': item['Province'],  # ÁúÅ‰ªΩ
            'City': item['City'],  # ÂüéÂ∏Ç
            'Signature': item['Signature'].replace('\n', ' ').replace(',', ' '),  # ‰∏™ÊÄßÁ≠æÂêçÔºàÂ§ÑÁêÜÁ≠æÂêçÂÜÖÂÆπÊç¢Ë°åÁöÑÊÉÖÂÜµÔºâ
            'StarFriend': item['StarFriend'],  # ÊòüÊ†áÂ•ΩÂèãÔºö1ÊòØÔºå0Âê¶
            'ContactFlag': item['ContactFlag']  # Â•ΩÂèãÁ±ªÂûãÂèäÊùÉÈôêÔºö1Âíå3Â•ΩÂèãÔºå259Âíå33027‰∏çËÆ©‰ªñÁúãÊàëÁöÑÊúãÂèãÂúàÔºå65539‰∏çÁúã‰ªñÁöÑÊúãÂèãÂúàÔºå65795‰∏§È°πËÆæÁΩÆÂÖ®Á¶ÅÊ≠¢
        }
        print(friend)
        friends.append(friend)
    return friends

if __name__ == '__main__':
    print(parse_data(get_data()))
```

```python
def save_to_txt():
    friends = parse_data(get_data())
    for item in friends:
        with open('friends.txt', mode='a', encoding='utf-8') as f:
            f.write('%s,%s,%d,%s,%s,%s,%d,%d\n' % (
                item['NickName'], item['RemarkName'], item['Sex'], item['Province'], item['City'], item['Signature'],
                item['StarFriend'], item['ContactFlag']))

if __name__ == '__main__':
    save_to_txt()
```

```python
from pyecharts import Bar

# Ëé∑ÂèñÁâπÊÆäÂ•ΩÂèã
star_list = []  # ÊòüÊ†áÊúãÂèã
deny_see_list = []  # ‰∏çËÆ©‰ªñÁúãÊàëÁöÑÊúãÂèãÂúà
no_see_list = []  # ‰∏çÁúã‰ªñÁöÑÊúãÂèãÂúà

with open('friends.txt', mode='r', encoding='utf-8') as f:
    rows = f.readlines()
    for row in rows:
        # # Ëé∑ÂèñÂ•ΩÂèãÂêçÁß∞
        name = row.split(',')[1] if row.split(',')[1] != '' else row.split(',')[0]
        # Ëé∑ÂèñÊòüÊ†áÊúãÂèã
        star = row.split(',')[6]
        if star == '1':
            star_list.append(name)
        # Ëé∑ÂèñËÆæÁΩÆ‰∫ÜÊúãÂèãÂúàÊùÉÈôêÁöÑÊúãÂèã
        flag = row.split(',')[7].replace('\n', '')
        if flag in ['259', '33027', '65795']:
            deny_see_list.append(name)
        if flag in ['65539', '65795']:
            no_see_list.append(name)

print('ÊòüÊ†áÂ•ΩÂèãÔºö', star_list)
print('‰∏çËÆ©‰ªñÁúãÊàëÁöÑÊúãÂèãÂúàÔºö', deny_see_list)
print('‰∏çÁúã‰ªñÁöÑÊúãÂèãÂúàÔºö', no_see_list)

attr = ['ÊòüÊ†áÊúãÂèã', '‰∏çËÆ©‰ªñÁúãÊàëÁöÑÊúãÂèãÂúà', '‰∏çÁúã‰ªñÁöÑÊúãÂèãÂúà']
value = [len(star_list), len(deny_see_list), len(no_see_list)]

bar = Bar('ÁâπÊÆäÂ•ΩÂèãÂàÜÊûê', '', title_pos='center')
bar.add('', attr, value, is_visualmap=True, is_label_show=True)
bar.render('ÁâπÊÆäÂ•ΩÂèãÂàÜÊûê.html')
bar
```
